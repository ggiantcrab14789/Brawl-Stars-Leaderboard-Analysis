```python
import brawlstats
import time
```

```python
client = brawlstats.Client('put your token here')
# Do not post your token on a public github
```

## Get Player Name

```python
player_name=[]
leaderboard = client.get_leaderboard('players', count=200)  # gets top 5 players
for player in leaderboard:
    player_name.append(player.name)
    
print(player_name)
```

## Get Player Tag

```python
player_tag=[]
leaderboard = client.get_leaderboard('players', count=200)  # gets top 5 players
for player in leaderboard:
    player_tag.append(player.tag)

print(player_tag)
```

## Get Player Exp

```python
player_exp=[]
leaderboard = client.get_leaderboard('players', count=200)
for player in leaderboard:
    player_exp.append(player.exp_level)


len(player_exp)
print(player_exp)
```

## Get Player Trophies

```python
player_trophies=[]
leaderboard = client.get_leaderboard('players', count=200)  # gets top 5 players
for player in leaderboard:
    player_trophies.append(player.trophies)

print(player_trophies)
```

## Get Player 3v3 victories

```python
import time
player_victories=[]

for k in range(200):
    client = brawlstats.Client('put your token here)
    info = client.get_profile(player_tag[k])
    player_victories.append(info.victories)
    time.sleep(1)

print(player_victories)
len(player_victories)
```

## Get Player Solo Showdown victories

```python
player_solo_showdown_victories=[]

for k in range(200):
    client = brawlstats.Client('put your token here)
    info = client.get_profile(player_tag[k])
    player_solo_showdown_victories.append(info.solo_showdown_victories)
    time.sleep(1)

print(player_solo_showdown_victories)
len(player_solo_showdown_victories)
```

## Get Player Duo Showdown victories

```python
player_duo_showdown_victories=[]

for k in range(200):
    client = brawlstats.Client('put your token here)
    info = client.get_profile(player_tag[k])
    player_duo_showdown_victories.append(info.duo_showdown_victories)
    time.sleep(1)

print(player_duo_showdown_victories)
len(player_duo_showdown_victories)
```

## Get Player's Club

```python
player_club_name=[]

for k in range(200):
    time.sleep(1)
    client = brawlstats.Client('put your token here)
    info = client.get_profile(player_tag[k])
    if info.club is None:
        info.club ='No Club'
        player_club_name.append(info.club)
    else:
        player_club_name.append(info.club.name)
        
print(player_club_name)
len(player_club_name)
```

## Get Player's Top Brawlers - Rank 1 to 3

```python
R1_brawler_name=[]
R1_brawler_trophies=[]
R2_brawler_name=[]
R2_brawler_trophies=[]
R3_brawler_name=[]
R3_brawler_trophies=[]
for k in range(200):
    client = brawlstats.Client('put your token here)
    info = client.get_profile(player_tag[k])
    R1_brawler_name.append(info.brawlers[0].name)
    R1_brawler_trophies.append(info.brawlers[0].trophies)
    #time.sleep(1)
    R2_brawler_name.append(info.brawlers[1].name)
    R2_brawler_trophies.append(info.brawlers[1].trophies)
    #time.sleep(1)
    R3_brawler_name.append(info.brawlers[2].name)
    R3_brawler_trophies.append(info.brawlers[2].trophies)
    time.sleep(1)

print(R1_brawler_name)
print(R1_brawler_trophies)
print(R2_brawler_name)
print(R2_brawler_trophies)
print(R3_brawler_name)
print(R3_brawler_trophies)
```
